The real-space Wannier Hamiltonian obtained by Wannier90 code generally does not strictly preserve the corresponding symmetry. The slight loss of symmetry will gap the expected band intersections. Berry curvature and orbital magnetic moment are more sensitive to symmetry than energy eigenvalues. In the following, I give details about the symmetrization of Wannier Hamiltonian and matrices.

In WannierBerri, our built-in symmetrization method can enforce symmetrizing Wannier Hamiltonian and matrices after reading them.

NOTE: This method only works with Wannier Hamiltonian without maximal localized. 

We will show you two examples:
1. Symmetrize nonmagnetic system with SOC (trigonal Te)
2. Symmetrize magnetic system with SOC (bcc Fe)

```python
#import necessary packages
import wannierberri as wberri
import numpy as np
import matplotlib.pyplot as plt #for plotting figures
```

## Nonmagnetic system (trigonal Te)

```python
#Reading Wannier Hamiltonian and position element matrix from Te_tb.dat file.
#We bulit to system, `system_Te` without symmetrization and `system_Te_sym` with symmetrization

system_Te=wberri.System_tb(tb_file='Te_tb.dat',berry=True,use_ws=False)
system_Te_sym=wberri.System_tb(tb_file='Te_tb.dat',berry=True,use_ws=False)

#Symmetrization step.
system_Te_sym.symmetrize(
    # positions: positions of each atoms in a unitcell (direct coordinate system)
    positions = np.array([[0.274, 0.274, 0.0],   
                    [0.726, 0.0, 0.33333333],
                   [0.0, 0.726, 0.66666667]]),
    # atom_name: name of each atoms, need to correspond to position order.
    atom_name = ['Te','Te','Te'],
    # projections: should be the same with projection card in wannier90.win
    proj = ['Te:s','Te:p'],
    # soc: with soc or spinless
    soc=True,
    # DFT_code: What code was used to complet the DFT calculation. only surport VASP and QE now.
    DFT_code='vasp',
    )
```

The Brillouin zone of trigonal Te is:
<img src="./Te_BZ.png" width="300">
We only consider kz=$\pi$ plane. 
<img src="./Te_kz=pi.png" width="300">
The band structures of the three path:

L1 -> H -> A1

L2 -> H -> A2

L3 -> H -> A3

should be the same because of C3 rotation symmetry.

But the symmetry of our Wannier Hamitonian brake the symmetries slightly. So the band structure of the three path may not the same.

After symmetrization they are the same.


```python
#generate k-path
path=wberri.Path(system_Te,
    #start and end points of k_path.
    #If the two points are not consencutive, put `None` between them. 
    k_nodes=[[0,0.5,0.5],[1./3,1./3,0.5],[1.,0.,0.5],None, 
            [0.5,0.,0.5],[1./3,1./3,0.5],[0.,1.,0.5],None,
            [0.5,0.5,0.5],[1./3,1./3,0.5],[0.,0.,0.5]],
    labels=["L1","H","G1","L2","H","G2","L3","H","G2"],
    #length: proportional to number of kpoints.
    length=1000) 

#calculate band structure with unsymmetrized and symmetrized system.
path_result_Te=wberri.tabulate(system_Te,
                 grid=path,
                 )
path_result_Te_sym=wberri.tabulate(system_Te_sym,
                 grid=path,
                 )
```

```python
#read energy eigenvalue data of band 18 and 19 from results.
band_Te = path_result_Te.get_data(quantity='Energy',iband=(18,19))
band_Te_sym = path_result_Te_sym.get_data(quantity='Energy',iband=(18,19))
#read k-path data
band_k=path.getKline()
```

```python
#plot band structure with unsymmetrized system
NK=len(band_k)
colour = ['r','y','b']
NK_one = NK//3
plt.figure()
for i in range(3):
    plt.plot(band_k[0:NK_one],band_Te[NK_one*i:NK_one*(i+1)],colour[i],linewidth=5)
plt.ylabel('Energy (eV)')
plt.xlabel('L -> H -> A')
```

```python
#plot band structure with symmetrized system
plt.figure()
for i in range(3):
    plt.plot(band_k[0:NK_one],band_Te_sym[NK_one*i:NK_one*(i+1)],colour[i],linewidth=5)
plt.ylabel('Energy (eV)')
plt.xlabel('L -> H -> A')
```

Symmetrization fix the symmetry of Hamiltonian

And position element matrix is symmetrized as well. It was used to calculate Berry curvature.

```python
# calculate Berry curvature alone a k-path.
path_berry=wberri.Path(system_Te,
                 k_nodes=[[0,0.5,0.5],[1./3,1./3,0.5],[1./3,1./3,0.]],
                 labels=["L","H","K",],
                 length=500)
berry_result_Te=wberri.tabulate(system_Te,
                 grid=path_berry,
                 quantities=['berry']
                 )
berry_result_Te_sym=wberri.tabulate(system_Te_sym,
                 grid=path_berry,
                 quantities=['berry']
                 )
```

```python
# use built-in plotting method to plot berry cuvature.
# unsymmetrized:
berry_result_Te.plot_path_fat(path_berry,
              quantity='berry',
              component='z', # only take z compoment
              Emin=6.0,  Emax=6.8,
              iband=[18,19],
              mode="fatband",
              fatfactor=20,  # size of dots.
              )
```

```python
# symmetrized: 
berry_result_Te_sym.plot_path_fat(path_berry,
              quantity='berry',
              component='z',
              Emin=6.0,  Emax=6.8,
              iband=[18,19],
              mode="fatband",
              fatfactor=20,
              )
```

z-component of Berry curvature should be zero because of symmetry. A slight loss of symmetry has a big impact to Berry curvature of single k-point. It also can be fix by symmetrization.

## Magnetic system with SOC (bcc Fe)

bcc Fe is a magnetic system with magnetic moment -2.31 \mu_B point to -z direction of cartesian coordinate system of lattice. We need to set parameter 'magmom'.

```python
system_Fe=wberri.System_tb(tb_file='Fe_sp3d2_tb.dat',berry=True,use_ws=False)
system_Fe_sym=wberri.System_tb(tb_file='Fe_sp3d2_tb.dat',berry=True,use_ws=False)
system_Fe_sym.symmetrize(
             proj = ['Fe:sp3d2;t2g'],
             atom_name = ['Fe'],
             positions = [[0,0,0]],
             #magmom: magnetic moment of each atoms.
             magmom = [[0.,0.,-2.31]],
             soc=True,
             DFT_code = 'qe'
             )
```

Important explanations:
1. The projection card in wannier90.win of bcc Fe is `'Fe':sp3d2;dxz,dyz,dxy`. But in the symmetrization, orbitals must project to complete sets of coordinates after symmetry opration. So we must label orbitals which shares the same complet sets of coordinates. eg: `sp3d2, dxz,dyz,dxy -> sp3d2, t2g` and `sp,px,py -> sp,p2`
2. We print out all the symmetry operators in the space group. But magnetic moment break some of them. Afte line `Symmetrizing start`. You can find detals about which operator belong to the magnetic group.

bcc Fe have inversion symmetry, so the Berry curvature dipole should be zero.

```python
#calculate berry curvature dipole
Efermi = Efermi=np.linspace(16.7,17.2,301)
wberri.run(system_Fe,
            grid=wberri.Grid(system_Fe,NK=10),
            calculators = {"BD":wberri.calculators.static.BerryDipole_FermiSea(Efermi=Efermi,tetra=True)},
            adpt_num_iter=0,
            fout_name='Fe',
            restart=False,
            )
wberri.run(system_Fe_sym,
            grid=wberri.Grid(system_Fe_sym,NK=10),
            calculators = {"BD":wberri.calculators.static.BerryDipole_FermiSea(Efermi=Efermi,tetra=True)},
            adpt_num_iter=0,
            fout_name='Fe_sym',
            restart=False,
            )
```

```python
#read data
Fe_BD = np.loadtxt("Fe-BD_iter-0000.dat")
Fe_sym_BD = np.loadtxt("Fe_sym-BD_iter-0000.dat")
f = open('Fe-BD_iter-0000.dat', "r")
title = f.readline().split()
f.close()
```

```python
#unsymmetrized
#print yx component of berry dipole
fig=plt.figure()
fig.suptitle(title[5], fontsize=20)
plt.plot(Fe_BD[:,0],Fe_BD[:,4],'r') 
plt.ylim(-0.00015,0.0003)
plt.xlabel('Energy (eV)')
plt.ylabel('Berry curvature dipole')
```

```python
#unsymmetrized
#print yx component of berry dipole
fig=plt.figure()
fig.suptitle(title[5], fontsize=20)
plt.plot(Fe_sym_BD[:,0],Fe_sym_BD[:,4],'r') 
plt.ylim(-0.00015,0.0003)
plt.xlabel('Energy (eV)')
plt.ylabel('Berry curvature dipole')
```

Symmetrized system give us beter result with low density k-grid.
